## Crear Regresiones

### Crear una función para agilizar las regresiones

```python
def regresion(df, x, y):
    model = linear_model.LinearRegression()
    X = df[x].values.reshape(-1, 1)
    Y = df[y]
    
    # Ajustar el modelo
    model.fit(X, Y)
    b0 = model.intercept_
    b1 = model.coef_[0]
    mse = mean_squared_error(y_true=Y, y_pred=model.predict(X))
    r2 = r2_score(y_true=Y, y_pred=model.predict(X))
    
    # Imprimir resultados
    print()
    print()
    print()
    print()
    print()
    print(f'----Resultados de la Regresión ({x[0]}):----')
    print()
    print(f'B0: {b0:.4f}')
    print(f'B1: {b1:.4f}')
    print(f'MSE: {mse:.4f}')
    print(f'R2: {r2:.4f}')
    print()
    
    # Graficar resultados
    plt.figure(figsize=(12, 8))
    plt.scatter(df[x], df[y], label='Datos')
    plt.scatter(df[x], model.predict(X), color='red', label='Regresión')
    plt.title(f'Regresión utilizando {x[0]}')
    plt.xlabel(x[0])
    plt.ylabel(y)
    plt.grid()
    plt.legend(loc='best')
    plt.show()
```

### Revisar si hay valores nulos en los datos 

```python
df.isna().mean()
```

Dado que la variable `total_bedrooms` tiene valores nulos, hay que eliminarlos para poder correr las regresiones.

```python
df2 = df.dropna()
```

### Revisar las variables que se usarán

```python
df.keys()
```

```python
x = ['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households',
     'median_income', 'near_bay', 'less_1hour_ocean', 'inland', 'near_ocean', 'island']
```

### Correr las regresiones 

```python
for variable in x:
    regresion(df2, [variable], 'median_house_value')
```

## Conclusiones

+ La regresión con mejor ajuste fue en la que se utilizó la variable `median_income`, obteniendo un R2 de $0.4131$.
+ Por otro lado, la segunda variable con mejor ajuste fue una variable dummy. Esto quiere decir que hubo un variable que con solo `True` y `False`, explicó mejor los datos que otras las demás variables numéricas. `inland` obtuvo un R2 de $0.2489$.
+ Los resultados de las demás regresiones son demasiado malos por si solos, sin embargo, podría ser buena idea utilizarlas al realizar una regresión múltiple en el futuro.
